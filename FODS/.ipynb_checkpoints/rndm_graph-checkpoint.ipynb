```python
import networkx as nx
import random 
import matplotlib.pyplot as plt

prob=[]
for i in range(10):
    p=random.random()
    prob.append(p)
    
prob=sorted(prob)

for i in range(len(prob)):
    G=nx.erdos_renyi_graph(5,i)
    nx.draw(G,with_labels=True)
    print(f"Graph with p = {prob[i]}")
    plt.show()
```
